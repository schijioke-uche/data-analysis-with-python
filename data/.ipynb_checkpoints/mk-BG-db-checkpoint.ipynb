```python
# Get also a normally distributed random sample:
from scipy.stats import norm
import numpy as np

# Get random samples
n = 40

# Get random numbers, normally distributed, centered at 100, std=15
A = norm.rvs(size=n,loc=110,scale=10)
A = np.sort(A)
A = A.reshape((n,1))


# Get random numbers, normally distributed, centered at 95, std=10
B = norm.rvs(size=n,loc=108,scale=8)
B = np.sort(B)
B = B.reshape((n,1))

# Get random number, normally distributed, centered at 90, std=5
C = norm.rvs(size=n,loc=106,scale=8)
C = np.sort(C)
C = C.reshape((n,1))
```

```python
AA = np.concatenate((A,B,C),axis=1)
AA
```

```python
import pandas as pd
df = pd.DataFrame.from_dict(AA)
```

```python
df.columns=['BG 1', 'BG 2','BG 3']
df.index.name='Person ID'
df
```

```python
df.to_csv('BG-db.csv')
```
